```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 
```

```python
df = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(df).transform(df)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
```

```python
df = pd.read_csv ('Market_Basket_Optimisation.csv')
df.head(5)
```

```python
records = []
for i in range(0, len(df)):
    records.append([str(df.values[i,j]) for j in range(0, 20)])
    records[i] = [x for x in records[i] if x != 'nan']
```

```python
te=TransactionEncoder()
te_ary=te.fit(records).transform(records)    
df1=pd.DataFrame(te_ary, columns=te.columns_)  
df1.head()
```

```python
frequent_itemsets=apriori(df1, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3)
```
